# Calculations

```python
def calculateScores(teams, week, bakers, dancers, leaderboard, spotlights):

    #12 players

    # for each player
    for player in teams.index:

        #get their bakers and dancers
        bo_team = teams.loc[player][:3]
        dwts_team = teams.loc[player][3:]

        baker_scores = 0
        dancer_scores = 0
        last_weeks_score = 0
        this_weeks_score = 0
        total_score = 0

        if week != 1:
            last_weeks_score = int(leaderboard.loc[player]["Total"])

        # for each baker on team, add points to participant's total score
        for x in bo_team:

            #baker 1, 2, 3
            for baker in bakers.index:
                if baker == x:
                    if week == 6 and spotlights.loc[player][0] == baker:
                      baker_scores += int(bakers.loc[baker]["Week " + str(week)]) * 2
                    else:
                      baker_scores += int(bakers.loc[baker]["Week " + str(week)])

        total_score = last_weeks_score + baker_scores

        # for each dancer on team, add points to the participant's total score
        for y in dwts_team:

            #dancer 1, 2, 3
            for dancer in dancers.index:
                if dancer == y:
                    if week == 6 and spotlights.loc[player][1] == dancer:
                      dancer_scores += int(dancers.loc[dancer]["Week " + str(week)]) * 3
                    else:
                      dancer_scores += int(dancers.loc[dancer]["Week " + str(week)])

        total_score += dancer_scores
        this_weeks_score = baker_scores + dancer_scores

        leaderboard.at[player,"Week " + str(week)] = this_weeks_score
        leaderboard.at[player,"Total"] = total_score

```

```python
calculateScores(players, 1, bakers, dancers, leaderboard,empty_spotlights)
calculateScores(players, 2, bakers, dancers, leaderboard, empty_spotlights)
calculateScores(players, 3, bakers, dancers, leaderboard, empty_spotlights)
calculateScores(players, 4, bakers, dancers, leaderboard, empty_spotlights)
calculateScores(players, 5, bakers, dancers, leaderboard, empty_spotlights)
calculateScores(swapped_players, 6, bakers, dancers, leaderboard, spotlights)
calculateScores(swapped_players, 7, bakers, dancers, leaderboard, empty_spotlights)
calculateScores(swapped_players, 8, bakers, dancers, leaderboard, empty_spotlights)
calculateScores(swapped_players, 9, bakers, dancers, leaderboard, empty_spotlights)
calculateScores(swapped_players, 10, bakers, dancers, leaderboard, empty_spotlights)

```
